## 03 The big merge

```python
orders_products_merged = orders_products_combined.merge(df_prods, on = 'product_id', how = 'outer', indicator = True)
```

## 04 Post-merge checks

```python
orders_products_merged.head()
```

```python
# used and outer join so that we can actually see the _merge column flag which things are not found in both dfs
orders_products_merged['_merge'].value_counts()
```

```python
# these orders listed product numbers that were not listed in the products list
df_left_only = orders_products_merged[orders_products_merged['_merge'] == 'left_only']
df_left_only
```
